Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Scatter plot')

# Show the plot
plt.show()
```

Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30)

# Add labels and title
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.title('Histogram')

# Show the plot
plt.show()
```

Fruit Quantities

```python
import matplotlib.pyplot as plt

# Define data
x = ['Apples', 'Oranges', 'Bananas']
y = [25, 15, 30]

# Create a bar chart
plt.bar(x, y)

# Add labels and title
plt.xlabel('Fruit')
plt.ylabel('Quantity')
plt.title('Fruit Quantities')

# Show the plot
plt.show()
```

Seaborn

```python
import seaborn as sns
import pandas as pd

# Load dataset
flights = sns.load_dataset('flights')
flights = flights.pivot('month', 'year', 'passengers')

# Create a heatmap
sns.heatmap(flights, annot=True, fmt='d')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Numbers')

# Show the plot
plt.show()
```

Pie Chart

```python
import matplotlib.pyplot as plt

# Define data
labels = ['Apples', 'Oranges', 'Bananas']
sizes = [25, 15, 30]

# Create a pie chart
plt.pie(sizes, labels=labels)

# Add title
plt.title('Fruit distribution')

# Show the plot
plt.show()
```

Box Plot

```python
#9. Box Plot
#9.1. Creating datas
data1 = np.random.normal(100, 10, 100)
data2 = np.random.normal(80, 30, 100)
data3 = np.random.normal(60, 50, 100)
     

#9.2. Combining datas
data = [data1, data2, data3]
     

#9.3. Plotting box plot
figure = plt.figure(figsize=(8, 5))
ax = figure.add_axes([0, 0, 1, 1])
ax.boxplot(data)
plt.show()
```

3D Scatter Plot

```python
#10. scatter Plot
# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt


# Creating dataset
z = np.random.randint(100, size =(50))
x = np.random.randint(80, size =(50))
y = np.random.randint(60, size =(50))

# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(x, y, z, color = "green")
plt.title("simple 3D scatter plot")

# show plot
plt.show()
```

3D Surface Plot

```python
#11. 3D Surface plot
#11.1. Creating datas
x = np.outer(np.linspace(-5,5,25), np.ones(25))
y = x.copy().T
z = (np.sin(x**2) + np.cos(y**2))
     

#11.2. Plotting 3D surface
figure = plt.figure(figsize = (10, 7))
ax = plt.axes(projection = '3d')
ax.plot_surface(x, y, z)
plt.show()
```
